```python
Y=df2.iloc[:,-1]
X=df2.iloc[:,:-1]
```

### 80% Train Data and 20% Test Data

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=0.2) ## Splitting the train and test model, so trained model is 80% and test model is 20%
model.fit(x_train, y_train) 
```

```python
y_predict=model.predict(x_test)
```

```python
confusion_matrix(y_predict,y_test)
```

### Accuracy score by applying Logistic Regression is 99.877%

```python
accuracy_score(y_test,y_predict)*100
```

### 160 Regression in the Neural Model 

```python
model=MLPClassifier(hidden_layer_sizes=(100,60))
model.fit(x_train, y_train)
```

```python
y_predict=model.predict(x_test)
confusion_matrix(y_predict,y_test)
```

### The accuracy score by applying Neural Model is 100%
